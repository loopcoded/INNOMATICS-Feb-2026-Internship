1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    status = "Full" if parked >= capacity else "Available"

    print(f"Currently Parked Vehicles: {parked}")
    print(f"Parking Status: {status}")

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, logs)
```

2: Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    time = distance * 5

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8

    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print(f"Estimated Delivery Time: {time} minutes")
delivery_time_estimator(8, "High", "Rainy")
```

3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Available"

    print(f"Occupancy: {int(occupancy)}%")
    print(f"Show Status: {status}")
booked_seats = [1] * 150
seat_occupancy_analyzer(200, booked_seats)
```

4: Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {int(avg_cpu)}%")
    print(f"Server Status: {status}")
server_load_classifier([45, 60, 70, 85, 90])
```

5: Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert: Yes" if overused else "Energy Alert: No")
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
classroom_resource_monitor(resources)
```

6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
event_registration_controller(100, 105)
```
