Unique Words in a Sentence

```python
sentence = "THIS IS SEJAL SINHA"
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Highest Salary from Employee Data

```python
employees = {
    "Sejal": 75000,
    "Kumud": 68000,
    "Yuktha": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
max_value = numbers[0]
min_value = numbers[0]
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_days = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_days / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

Count Character Frequency

```python
text = "python"
char_frequency = {}

for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print(char_frequency)
```

Check if a Key Exists in a Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000
}

if "Ravi" in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Calculate Average Marks

```python
marks = [70, 75, 80, 65]

average_marks = sum(marks) / len(marks)

print("Average Marks:", average_marks)
```
