# Dictionary containing employee names and performance scores

```python
employees = {
    "sejal": 92,
    "shagun": 88,
    "Kiran": 92,
    "Suresh": 85
}
# highest performance score
highest_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == highest_score]
# Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

#2: Search Query Keyword Analysis

```python
import string

query = "love india , LOVE COUNTRY"

# Convert to lowercase
query = query.lower()

# Remove punctuation
for char in string.punctuation:
    query = query.replace(char, "")

# Split into words
words = query.split()

# Count frequency
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

repeated_keywords = {word: count for word, count in word_count.items() if count > 1}

print(repeated_keywords)
```

#3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Iterate through sensor readings with index
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

#4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract and count email domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

# Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

#5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Detect sales spikes (> 30% above average)
threshold = average_sales * 1.3

for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

#6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

# Count occurrences of user IDs
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Display duplicates only
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
